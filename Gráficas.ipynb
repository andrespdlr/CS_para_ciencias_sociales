```python
import random
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
X = []
Y = []

for i in range(20):
    x = random.randint(0, 100)
    y = random.randint(0, 100)
    X.append(x)
    Y.append(y)
```

```python
from matplotlib.patches import Rectangle
```

```python
import matplotlib.mlab as mlab
from matplotlib.path import Path
from matplotlib.spines import Spine
from matplotlib.projections.polar import PolarAxes
from matplotlib.projections import register_projection

from matplotlib.patches import Polygon
from pandas.tools.plotting import table
import colorsys
import textwrap
```

```python
fig = plt.figure(figsize=(12, 12))
plt.tick_params(axis='both', which='major', labelsize=15)
plt.title("Valores Aleatorios", fontsize = 20, fontweight = 'bold')
plt.xlabel("\nValor X", fontsize = 15)
plt.ylabel("Valor Y", fontsize = 15)
plt.ylim(0, 100)
xs_ticks = np.arange(len(X))
plt.xticks(xs_ticks, X)
plt.bar(xs_ticks, Y, color = 'blue', align = 'center', edgecolor = 'black')
plt.savefig('chart_belisario_1.png', dpi = 100, bbox_inches="tight", bbox_extra_artists=[])
```

```python
mean_X = sum(X) / len(X)
mean_Y = sum(Y) / len(Y)
fig = plt.figure(figsize=(12, 12))
axis = plt.gca()
plt.tick_params(axis='both', which='major', labelsize=15)
plt.title("Valores Aleatorios", fontsize = 20, fontweight = 'bold')
plt.xlabel("Valor X", fontsize = 15)
plt.ylabel("Valor Y", fontsize = 15)
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.axvline(mean_X, color = 'black', linestyle = 'solid', linewidth = 2)
plt.axhline(mean_Y, color = 'black', linestyle = 'solid', linewidth = 2)

axis.add_patch(Rectangle((mean_X, mean_Y), 100 - mean_X, 100 - mean_Y, facecolor = 'blue', alpha = 1))

plt.plot(X, Y, 'o', c = 'black', markersize = 5, alpha = .5)
plt.plot(75, 75, 'D', c = 'black', markersize = 10)
plt.text(80, 80, 'Valor', color = 'white', ha='center', bbox=dict(facecolor = 'black', alpha = 1), fontsize = 20)
plt.savefig('chart_belisario_2.png', dpi = 100, bbox_inches="tight", bbox_extra_artists=[])
```
