```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
plt.style.use('ggplot')
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
world.head()
```

```python
world.plot()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax.set_aspect('equal')
world.plot(ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax.axis('off')
ax.set_aspect('equal')
world.plot(ax=ax)
plt.show()
```

```python
world = world[(world.pop_est > 0) & (world.name != "Antarctica")]
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax.axis('off')
ax.set_aspect('equal')
world.plot(ax=ax)
plt.show()
```

```python
world.head()
```

```python
world['gdp_per_cap'] = world.gdp_md_est / world.pop_est
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax.axis('off')
ax.set_aspect('equal')
world.plot(ax=ax, column='gdp_per_cap')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax.axis('off')
ax.set_aspect('equal')
world.plot(ax=ax, column='gdp_per_cap', scheme='quantiles')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax.axis('off')
ax.set_aspect('equal')
world.plot(ax=ax, column='gdp_per_cap', scheme='equal_interval')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax.axis('off')
ax.set_aspect('equal')
world.plot(ax=ax, column='gdp_per_cap', scheme='fisher_jenks')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax.axis('off')
ax.set_aspect('equal')
world.plot(ax=ax, column='gdp_per_cap', cmap='Blues', scheme='quantiles')
plt.show()
```
