```python
asia = world[(world.continent == "Asia")]
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax.axis('off')
ax.set_aspect('equal')
asia.plot(ax=ax)
plt.show()
```

```python
stan_countries = [pais for pais in asia.name if 'stan' in pais[-4:]]
```

```python
stan_countries
```

```python
stan_condition = []
for row in range(len(world)):
    country = world['name'][row]
    if country in stan_countries:
        stan_condition.append('is_stan')
    else:
        stan_condition.append('not_stan')
asia['stan_condition'] = stan_condition
```
