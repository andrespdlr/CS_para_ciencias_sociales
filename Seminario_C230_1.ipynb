# Si contamos a todos en la tierra, cuántos años tardaríamos?

```python
total_personas = 7000000000
chunks = 2
segundos_por_cuenta = 1
```

```python
total_cuentas = total_personas / chunks
segundos = total_cuentas * segundos_por_cuenta
minutos = segundos / 60
horas = minutos / 60
dias = horas / 24
años = round(dias / 365, 2)
print('Contar', total_personas, 'personas', 'de {} en {}'.format(chunks, chunks), 'tomaría {} '.format(años), 'años')
print('El número final de pasos necesarios es : ', total_cuentas)
```

# ... y si los contamos correctamente?

```python
numero_de_pasos = 0
while total_personas != 1:
    if total_personas%2 == 0:
        total_personas = total_personas / 2
    else:
        total_personas = (total_personas + 1) / 2
    numero_de_pasos += 1
print('El número final de pasos necesarios es : ', numero_de_pasos)
```

## Qué tal que necesitamos determinar si un número es primo

```python
for n in range(2, 10):
    is_prime = 'es primo'
    for i in range(2, n):
        if n % i == 0:
            is_prime = 'no es primo'
            break
    print(n, is_prime)
```

```python
def es_primo(numero):
    evaluacion = True
    for i in range(2, numero):
        if numero % i == 0:
            evaluacion = False
            break
    return evaluacion
```

```python
def primos_antes_de_numero(numero):
    rv = []
    for i in range(2, numero):
        if es_primo(i) == True:
            rv.append(i)
        else:
            pass
    return rv
```

```python
def primos_antes_de_numero_fea(numero):
    rv = []
    for i in range(2, numero):
        evaluacion = True
        for j in range(2, i):
            if i % j == 0:
                evaluacion = False
                break
        if evaluacion == True:
            rv.append(i)
    return rv
```

```python
def sector_SCIAN(rama):
    '''
    Converts rama into sector
    
    Input: rama scian
    
    Output: sector scian
    '''
    sector = rama[0: 2]
    if sector == '31' or sector == '32' or sector == '33':
        rv = '31-33'
    elif sector == '48' or sector == '49':
        rv = '48-49'
    else:
        rv = sector
    return rv
```
